# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_path = ('Resources/whale_returns.csv')
whale_returns = pd.read_csv(whale_path)
whale_returns = whale_returns.set_index(pd.to_datetime(whale_returns['Date'], infer_datetime_format = True))
whale_returns.drop(columns=['Date'], inplace= True)
whale_returns.head()
```

```python
# Count nulls
whale_returns.isnull().sum()
```

```python
# Drop nulls
whale_returns = whale_returns.dropna()
whale_returns.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_path = ('Resources/algo_returns.csv')
algo_returns = pd.read_csv(algo_path)
algo_returns = algo_returns.set_index(pd.to_datetime(algo_returns['Date'], infer_datetime_format = True))
algo_returns.drop(columns=['Date'], inplace= True)
algo_returns.head()
```

```python
# Count nulls
algo_returns.isnull().sum()
```

```python
# Drop nulls
algo_returns = algo_returns.dropna()
algo_returns
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp_path = ('Resources/sp500_history.csv')
sp_df = pd.read_csv(sp_path)
sp_df = sp_df.set_index(pd.to_datetime(sp_df['Date'], infer_datetime_format = True))
sp_df.drop(columns = ['Date'], inplace= True)
sp_df['Close'] = sp_df['Close'].str.replace('$', '').astype('float64')
sp_df.head()
```

```python
# Check Data Types
sp_df.dtypes
```

```python
# Calculate Daily Returns
sp_returns = sp_df.pct_change()
sp_returns
```

```python
# Drop nulls
sp_returns.dropna()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_returns= sp_returns.rename(columns= {'Close': 'S&P Returns'})
sp_returns
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_df = pd.concat([whale_returns, algo_returns, sp_returns], axis = 'columns', join = 'inner')
combined_df.head()
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
weights = [.5,.5]
algo_cov = combined_df.loc[:, 'Algo 1':'Algo 2'].dot(weights).cov(combined_df['S&P Returns'])
algo_rolling_cov = combined_df.loc[:, 'Algo 1': 'Algo 2'].dot(weights).rolling(window=7).cov(combined_df['S&P Returns'])

# Calculate variance of S&P 500
variance = sp_returns.var()
rolling_variance = combined_df['S&P Returns'].rolling(window=7).var()

# Computing beta
algo_beta = algo_cov / variance

# Plot beta trend, window of 60
algo_rolling_beta = algo_rolling_cov / rolling_variance
algo_rolling_beta.plot(figsize=(20,10))
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
annual_std = combined_df.std() * np.sqrt(252)
sharpe_ratios = (combined_df.mean() *252) / annual_std
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot.bar()
```
